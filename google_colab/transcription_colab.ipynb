# Транскрибация аудио через Whisper

Этот ноутбук транскрибирует аудио файлы через API Video Maker сервиса.

## Как использовать:

1. Запустите все ячейки по порядку
2. Введите URL вашего сервера (например: http://your-server.com:5000)
3. Введите task_id из вашего workflow
4. Ноутбук скачает аудио, транскрибирует через Whisper и отправит результат обратно
